```python
import matplotlib.pyplot as plt
from fractions import Fraction

def primal_line_to_dual_point(p1, p2):
    x1, y1 = p1
    x2, y2 = p2

    if x2 == x1:
        raise ValueError("Vertical line has undefined slope; dual is not a point in standard duality.")
    
    m = (y2 - y1) / (x2 - x1)
    c = y1 - m * x1
    return (m, -c)

def orientation(p1, p2, p3):
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    
    # Using Fraction for exact rational arithmetic
    val = (Fraction(x2 - x1) * Fraction(y3 - y1)) - (Fraction(y2 - y1) * Fraction(x3 - x1))
    
    if val > 0:
        return "Counterclockwise"
    elif val < 0:
        return "Clockwise"
    else:
        return "Collinear"

# Blue
pl = (2, 2)
pi = (8, 9)

# Orange
pv = (3, 7)
pj = (10, 6)

# Green
pu = (3, 8)
pk = (9, 2)

# Compute dual points
p_li = primal_line_to_dual_point(pl, pi)
p_vj = primal_line_to_dual_point(pv, pj)
p_uk = primal_line_to_dual_point(pu, pk)

print(f"The orientation is {orientation(p_li, p_vj, p_uk)}")

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot primal lines
ax1.plot([pl[0], pi[0]], [pl[1], pi[1]], marker='o', color='blue')
ax1.plot([pv[0], pj[0]], [pv[1], pj[1]], marker='o', color='orange')
ax1.plot([pu[0], pk[0]], [pu[1], pk[1]], marker='o', color='green')
ax1.set_title("Primal Plane (Lines)")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.grid(True)
ax1.set_aspect('equal')

# Plot dual points


ax2.plot([p_li[0], p_vj[0]], [p_li[1], p_vj[1]], marker='o', color="black")
ax2.plot([p_li[0], p_uk[0]], [p_li[1], p_uk[1]], marker='o', color="black")

ax2.plot([p_li[0]], [p_li[1]], marker='o', color='blue')
ax2.plot([p_vj[0]], [p_vj[1]], marker='o', color='orange')
ax2.plot([p_uk[0]], [p_uk[1]], marker='o', color='green')

# ax2.plot([pv[0], pj[0]], [pv[1], pj[1]], marker='o', color='orange')
# ax2.plot([pu[0], pk[0]], [pu[1], pk[1]], marker='o', color='green')


ax2.set_title("Dual Plane (Points)")
ax2.set_xlabel("m (slope)")
ax2.set_ylabel("-c (negative y-intercept)")
ax2.grid(True)
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from fractions import Fraction

def primal_line_to_dual_point(p1, p2):
    denominator = p1[0] * p2[1] - p1[1] * p2[0]
    if denominator == 0:
        raise ValueError("Points are linearly dependent or degenerate (denominator = 0).")
    return (
        (p1[1] - p2[1]) / denominator, 
        (p2[0] - p1[0]) / denominator
    )

def orientation(p1, p2, p3):
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    
    # Using Fraction for exact rational arithmetic
    val = (Fraction(x2 - x1) * Fraction(y3 - y1)) - (Fraction(y2 - y1) * Fraction(x3 - x1))
    
    if val > 0:
        return "Counterclockwise"
    elif val < 0:
        return "Clockwise"
    else:
        return "Collinear"

# Blue
pl = (2, 2)
pi = (8, 9)

# Orange
pv = (2, 4)
pj = (10, 6)

# Green
pu = (6, 9)
pk = (9, 2)

# Teal
# pu2 = (3, 2)
# pk2 = (3, 5)

pu2 = (3, 5)
pk2 = (4, 2)

# Compute dual points
p_li = primal_line_to_dual_point(pl, pi)
p_vj = primal_line_to_dual_point(pv, pj)
p_uk = primal_line_to_dual_point(pu, pk)
p_u2k2 = primal_line_to_dual_point(pu2, pk2)

print(f"The orientation is {orientation(p_li, p_vj, p_uk)}")

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Primal lines
ax1.plot([pl[0], pi[0]], [pl[1], pi[1]], marker='o', color='blue', markersize=10)
ax1.plot([pv[0], pj[0]], [pv[1], pj[1]], marker='o', color='orange', markersize=10)
ax1.plot([pu[0], pk[0]], [pu[1], pk[1]], marker='o', color='green', markersize=10)
ax1.plot([pu2[0], pk2[0]], [pu2[1], pk2[1]], marker='o', color='teal', markersize=10)
# ax1.set_title("Primal Plane (Segments)")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_aspect('equal')
# ax1.set_xticklabels([])
# ax1.set_yticklabels([])
ax1.grid(True)

# Dual points
ax2.plot([p_li[0], p_vj[0]], [p_li[1], p_vj[1]], marker='o', color="black", linewidth=0.5)
ax2.plot([p_li[0], p_uk[0]], [p_li[1], p_uk[1]], marker='o', color="black", linewidth=0.5)
ax2.plot([p_li[0], p_u2k2[0]], [p_li[1], p_u2k2[1]], marker='o', color="black", linewidth=0.5)

ax2.plot([p_li[0]], [p_li[1]], marker='o', color='blue', markersize=10)
ax2.plot([p_vj[0]], [p_vj[1]], marker='o', color='orange', markersize=10)
ax2.plot([p_uk[0]], [p_uk[1]], marker='o', color='green', markersize=10)
ax2.plot([p_u2k2[0]], [p_u2k2[1]], marker='o', color='teal', markersize=10)

# ax2.set_title("Dual Plane (Affine RP² Coordinates)")
# ax2.set_xlabel("x (dual)")
# ax2.set_ylabel("y (dual)")
ax2.set_aspect('equal')
ax2.grid(True)
# ax2.set_xticklabels([])
# ax2.set_yticklabels([])
plt.tight_layout()
plt.show()
```

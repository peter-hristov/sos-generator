```python
from sympy import symbols, Matrix, IndexedBase, Idx, expand, collect, Indexed, latex, factor, simplify

from sympy import pprint

# Indexed bases for p and e
p = IndexedBase('p')
e = IndexedBase('e')

# Symbolic indices
i, j, k, l, u, v = symbols('i j k l u v')

s_li = Matrix([
    p[l, 2] - p[i, 2] + e[l, 2] - e[i, 2],
    p[i, 1] - p[l, 2] + e[i, 1] - e[l, 2],
    (p[i, 1] + e[i,1]) * (p[l, 2] + e[l, 2]) - (p[i, 2] + e[i, 2])*(p[l, 1] + e[l, 1])
])

s_vj = Matrix([
    p[v, 2] - p[j, 2] + e[v, 2] - e[j, 2],
    p[j, 1] - p[v, 2] + e[j, 1] - e[v, 2],
    (p[v, 1] + e[v,1]) * (p[j, 2] + e[j, 2]) - (p[v, 2] + e[v, 2])*(p[j, 1] + e[j, 1])
])

s_uk = Matrix([
    p[u, 2] - p[k, 2] + e[u, 2] - e[k, 2],
    p[k, 1] - p[u, 2] + e[k, 1] - e[u, 2],
    (p[u, 1] + e[u,1]) * (p[k, 2] + e[k, 2]) - (p[u, 2] + e[u, 2])*(p[k, 1] + e[k, 1])
])

# Convert to row vectors using transpose, then stack them
expression = expand(Matrix.vstack(s_li.T, s_vj.T, s_uk.T).det())
expressionTermsOrdered = expression.as_ordered_terms()

def count_indexed_with_base(term, base):
    """Return the number of Indexed objects in a term with the given base."""
    return sum(1 for atom in term.atoms(Indexed) if atom.base == base)


# e_terms = [t for t in espressionTermsOrdered if t.has(e)]
pure_p_terms = [t for t in expressionTermsOrdered if t.has(p) and not t.has(e)]
pure_e_terms = [t for t in expressionTermsOrdered if t.has(e) and not t.has(p)]
mixed_terms = [t for t in expressionTermsOrdered if t.has(e) and t.has(p)]


terms_1e_3p = [
    t for t in expressionTermsOrdered
    if count_indexed_with_base(t, e) == 1 and count_indexed_with_base(t, p) == 3
]

terms_2e_2p = [
    t for t in expressionTermsOrdered
    if count_indexed_with_base(t, e) == 2 and count_indexed_with_base(t, p) == 2
]

terms_3e_1p = [
    t for t in expressionTermsOrdered
    if count_indexed_with_base(t, e) == 3 and count_indexed_with_base(t, p) == 1
]

# pprint(f"There are {len(pure_p_terms)} pure p terms.")
# pprint(f"There are {len(pure_e_terms)} pure e terms.")
# pprint(f"There are {len(mixed_terms)} mixed  terms.")

# pprint(f"There are {len(terms_1e_3p)} 1e 3p terms.")
# pprint(f"There are {len(terms_2e_2p)} 2e 2p terms.")
# pprint(f"There are {len(terms_3e_1p)} 3e 1p terms.")

terms_pure_p_expr = sum(pure_p_terms)
terms_pure_e_expr = sum(pure_e_terms)
terms_mixed_expr = sum(mixed_terms)

terms_1e_3p_expr = sum(terms_1e_3p)
terms_2e_2p_expr = sum(terms_2e_2p)
terms_3e_1p_expr = sum(terms_3e_1p)

collectionTerm = e[k, 2]
collectionExpression = terms_1e_3p_expr

collected_expr = collect(collectionExpression, collectionTerm).coeff(collectionTerm)

print(latex(collected_expr))


pprint(f"Total terms {len(expressionTermsOrdered)}")
```

```python
from itertools import product
from sympy import S

def generate_sequence(varNames, dimensions=(1, 2)):
    sequence = []
    # sequence.append([f"e[{varNames[0]}, {dimensions[1]}]", f"e[{varNames[0]}, {dimensions[0]}]"])
    sequence.append([e[varNames[0], dimensions[1]], e[varNames[0], dimensions[0]]])

    for i in range(1, len(varNames)):
        currentSequence = []
        for j in reversed(range(len(dimensions))):
            # currentSequence.append(f"e[{varNames[i]}, {dimensions[j]}]")
            currentSequence.append(e[varNames[i], dimensions[j]])
            for s in range(len(sequence)):
                previousSequence = sequence[s]
                for k in range(len(previousSequence)):
                    # appendString = f"e[{varNames[i]}, {dimensions[j]}]"
                    appendProduct = e[varNames[i], dimensions[j]] * previousSequence[k]
                    # if previousSequence[k] != "":
                    # appendProduct *= previousSequence[k]
                    currentSequence.append(appendProduct)
        sequence.append(currentSequence)
                    
                    

    return sequence


# Example usage
# vars = ['i', 'j', 'k', 'l']

def print_terms():
    numberOfTerms = 0
    vars = symbols('i j k l u v')
    seq = generate_sequence(vars)
    for line in seq:
        for expression in line:
    
            
            # print(f"\n\nThe epsilon product with order {numberOfTerms}: {latex(expression, symbol_names={e: '\\varepsilon'})}")

            
            # print(expression)
            num_factors = len(expression.args) if expression.is_Mul else 1
            # # print(num_factors)
    
            collectionTerm = expression
    
            if (num_factors == 1):
                collectionExpression = terms_1e_3p_expr
            elif(num_factors == 2):
                collectionExpression = terms_2e_2p_expr
            elif(num_factors == 3):
                collectionExpression = terms_3e_1p_expr
            elif(num_factors == 4):
                collectionExpression = terms_pure_e_expr

            # Skip if it's not in the expression
            if not collectionExpression.has(collectionTerm):
                continue

            numberOfTerms+=1

            collected_expr = collect(collectionExpression, collectionTerm).coeff(collectionTerm)


            # Latex print
            # \hline
            # \raggedright
            # \(
            # - {p}_{i,1} + {p}_{u,1}
            # \) &
            # \({e}_{j,1} {e}_{k,2} {e}_{l,2}\)
            # \\

            print("\\hline")
            # print(f"{numberOfTerms} & ")
            print("\\raggedright")
            print("\\(")
            print(latex(collected_expr))
            print("\\) &")
            print(f"\\({latex(expression)}\\)")
            print("\\\\")

            # print(f"\n\nThe epsilon product with order {numberOfTerms}: {latex(expression)}")
            # print(f"The p-term of that product has {len(collected_expr.as_ordered_terms())} terms:")
            # print(latex(collected_expr))
            # print(f"Here it is simplified...")
            # print(simplify(collected_expr))
            
            

            # If there are four factors, there is no p in there
            if (num_factors == 4):
                print(f"There were a total of {numberOfTerms} terms")
                return

            # pprint(f"Total terms {len(expressionTermsOrdered)}")

print_terms()

# Define the indexed base and indices
# e = IndexedBase('e')
# i, j, k, l, m, n = symbols('i j k l m n')

# # Define a product expression
# term = e[i, j] * e[k, l] * e[m, n]

# # Count the number of multiplicative factors
# num_factors = len(term.args) if term.is_Mul else 1

# print(f"Number of factors: {num_factors}")
```

```python
import matplotlib.pyplot as plt

def primal_line_to_dual_point(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    
    if x2 == x1:
        raise ValueError("Vertical line has undefined slope; dual is not a point in standard duality.")
    
    m = (y2 - y1) / (x2 - x1)
    c = y1 - m * x1
    return (m, -c)

# Blue
pl = (2, 2)
pi = (8, 8)

# Orange
pv = (2, 4)
pj = (10, 6)

# Green
pu = (6, 8)
pk = (8, 2)


# Correct way to call plot: give x and y lists
plt.plot([pl[0], pi[0]], [pl[1], pi[1]], marker='o', color='blue')
plt.plot([pv[0], pj[0]], [pv[1], pj[1]], marker='o', color='orange')
plt.plot([pu[0], pk[0]], [pu[1], pk[1]], marker='o', color='green')

# Plot points
p_li = primal_line_to_dual_point(pl, pi)
p_vj = primal_line_to_dual_point(pv, pj)
p_uk = primal_line_to_dual_point(pu, pk)

plt.plot([p_li[0]], [p_li[1]], marker='o', color='blue', label='Dual Point')
plt.plot([p_vj[0]], [p_vj[1]], marker='o', color='orange', label='Dual Point')
plt.plot([p_uk[0]], [p_uk[1]], marker='o', color='green', label='Dual Point')



# Style the plot
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.title("Three Lines from Six Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

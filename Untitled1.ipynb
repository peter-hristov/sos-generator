```python
import matplotlib.pyplot as plt
import numpy as np
from fractions import Fraction
from sympy import Point, Line
from ipywidgets import interact, FloatSlider
import ipywidgets as widgets

def primal_line_to_dual_point(p1, p2):
    denominator = p1[0] * p2[1] - p1[1] * p2[0]
    if denominator == 0:
        raise ValueError("Points are linearly dependent or degenerate (denominator = 0).")
    return (
        (p1[1] - p2[1]) / denominator, 
        (p2[0] - p1[0]) / denominator
    )

def line_intersection(p1, p2, q1, q2):
    """
    Return the intersection point of the infinite lines through (p1, p2) and (q1, q2),
    as a tuple of floats. Returns None if lines are parallel or identical.
    """
    # Convert input tuples to sympy Points
    p1, p2 = Point(p1), Point(p2)
    q1, q2 = Point(q1), Point(q2)
    
    line1 = Line(p1, p2)
    line2 = Line(q1, q2)
    
    inter = line1.intersection(line2)
    
    if not inter:
        return None  # No intersection (parallel or coincident)
    
    pt = inter[0].evalf()
    return float(pt.x), float(pt.y)

def orientation(p1, p2, p3):
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    
    # Using Fraction for exact rational arithmetic
    val = (Fraction(x2 - x1) * Fraction(y3 - y1)) - (Fraction(y2 - y1) * Fraction(x3 - x1))
    
    if val > 0:
        return "Counterclockwise"
    elif val < 0:
        return "Clockwise"
    else:
        return "Collinear"

def draw_implicit_line(a, b, c, painter, xlim=(-0.9, 0.4), **kwargs):
    x = np.linspace(*xlim, 500)
    if b != 0:
        y = (-a * x - c) / b
        painter.plot(x, y, **kwargs)
    else:
        x0 = -c / a
        painter.axvline(x0, **kwargs)


def plot_lines(pl, pi, pv, pj, pu, pk):
    xMin = 0
    xMax = 10
    yMin = 0
    yMax = 10
    
    xdMin = -2
    xdMax = 2
    ydMin = -2
    ydMax = 2
        
    # Compute dual points
    p_li = primal_line_to_dual_point(pl, pi)
    p_vj = primal_line_to_dual_point(pv, pj)
    p_uk = primal_line_to_dual_point(pu, pk)

    
    print(f"The orientation is {orientation(p_li, p_vj, p_uk)}")
    
    # Create subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # Primal lines
    # Gray
    ax1.plot([pl[0], pi[0]], [pl[1], pi[1]], marker='o', color='gray', markersize=10, linewidth=3.5)
    # Orange
    ax1.plot([pv[0], pj[0]], [pv[1], pj[1]], marker='o', color='orange', markersize=10, linewidth=3.5)
    # Green
    ax1.plot([pu[0], pk[0]], [pu[1], pk[1]], marker='o', color='green', markersize=10, linewidth=3.5)
    
    
    ax1.plot([pi[0]], [pi[1]], marker='o', color='green', markersize=10)
    ax1.plot([pl[0]], [pl[1]], marker='o', color='blue', markersize=10)

    # ax1.plot([i_bo[0]], [i_bo[1]], marker='o', color='red', markersize=10)
    
    # ax1.set_title("Primal Plane (Segments)")
    ax1.set_xlabel("x")
    ax1.set_ylabel("y")
    ax1.set_aspect('equal')
    ax1.set_xlim(xMin, xMax)
    ax1.set_ylim(yMin, yMax)
    # ax1.set_xticklabels([])
    # ax1.set_yticklabels([])
    ax1.grid(True)



     # Dual points
    
    # Draw the bounds of the edges
    draw_implicit_line(pi[0], pi[1], 1, ax2, xlim = (-10, 10), color='green', linewidth=2.5)
    draw_implicit_line(pl[0], pl[1], 1, ax2, xlim = (-10, 10), color='blue', linewidth=2.5)
    
    # draw_implicit_line(i_bo[0], i_bo[1], 1, ax2, xlim = (-10, 10), color='gra', linewidth=0.5)
    
    # draw_implicit_line(pt[0], pt[1], 1, ax2, xlim = (-10, 10), color='red', linewidth=0.5)
    
    # draw_implicit_line(pu[0], pu[1], 1, color='gree', linewidth=0.5)
    # draw_implicit_line(pk[0], pk[1], 1, color='green', linewidth=0.5)
    
    
    
    
    
    
    
    
    
    
   
    
    
    
    # ax2.plot([p_li[0], p_vj[0]], [p_li[1], p_vj[1]], marker='o', color="black", linewidth=0.5)
    # ax2.plot([p_li[0], p_uk[0]], [p_li[1], p_uk[1]], marker='o', color="black", linewidth=0.5)
    # ax2.plot([p_li[0], p_u2k2[0]], [p_li[1], p_u2k2[1]], marker='o', color="black", linewidth=0.5)
    ax2.plot([p_li[0]], [p_li[1]], marker='o', color='gray', markersize=12)
    ax2.plot([p_vj[0]], [p_vj[1]], marker='o', color='orange', markersize=10)
    ax2.plot([p_uk[0]], [p_uk[1]], marker='o', color='green', markersize=10)
    
    # ax2.plot([i_bo_a[0]], [i_bo_a[1]], marker='o', color='teal', markersize=20)
    # ax2.plot([i_bo_b[0]], [i_bo_b[1]], marker='o', color='purple', markersize=20)
    
    
    # ax2.set_title("Dual Plane (Affine RPÂ² Coordinates)")
    # ax2.set_xlabel("x (dual)")
    # ax2.set_ylabel("y (dual)")
    ax2.set_aspect('equal')
    ax2.grid(True)
    ax2.set_xlim(xdMin, xdMax)
    ax2.set_ylim(ydMin, ydMax)
    # ax2.set_xticklabels([])
    # ax2.set_yticklabels([])
    plt.tight_layout()
    plt.show()

plot_lines(pl, pi, pv, pj, pu, pk)
```
